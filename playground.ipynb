```python
from torch.nn import Sequential
from dense.utils import RichRepr
from torch.nn import ReLU, BatchNorm2d, Conv2d, AvgPool2d, Flatten, Linear
class ClassificationBlock(RichRepr, Sequential):
    r"""
    Classification block for [DenseNet](https://arxiv.org/abs/1608.06993),
    takes in a 7x7 feature map and outputs logit scores for classification
    """

    def __init__(self, in_channels, output_classes, flat_nodes):
        super(ClassificationBlock, self).__init__()

        self.in_channels = in_channels
        self.out_classes = output_classes

        self.add_module('norm', BatchNorm2d(num_features=in_channels))
        self.add_module('relu', ReLU(inplace=True))
        self.add_module('pool', AvgPool2d(kernel_size=2, stride=1, padding=1))
        self.add_module('flatten', Flatten())
        self.add_module('linear', Linear(flat_nodes, output_classes))

    def __repr__(self):
        return super(ClassificationBlock, self).__repr__(self.in_channels, self.out_classes)

import pickle
with open('x_cat.pickle', 'rb') as file_in:
    inputs = pickle.load(file_in)

print(inputs.size())
class_thing = ClassificationBlock(52, 2, 936)
class_thing(inputs).size()
```
